```python
import pandas as pd
```

```python
df1 = pd.read_csv('1.csv')
df1.head()
```

```python
df2 = pd.read_csv('2.csv')
df2 = df2.rename(columns={'Country/Other': 'country'})
df2.head()
```

```python
df_merged = pd.merge(df1, df2, on='country', how='inner')
```

```python
df_merged.info()
```

```python
df = df_merged[['country','currency','capital_city','region','continent','gdp','agricultural_land','forest_area','land_area','inflation','co2_emissions','health_expenditure_capita','hospital_beds','suicide_rate','birth_rate','death_rate','fertility_rate','internet_pct','life_expectancy','population_female','population_male','population','rural_population','urban_population','democracy_type','median_age','Density (P/Km²)','Land Area (Km²)','World Share']]
```

```python
df.to_csv('dataset.csv',index=False)
```
